# MATPLOTLIB 
### Basics of graph types

```python
import numpy as np
import matplotlib.pyplot as plt
#####1
x1 = np.array([1,2,3,4,5])
y1 = np.array([10,13,15,8,4])
# plt.bar(x1,y1, color = "red", width="0.5")
plt.bar(x1,y1, color= "green")
plt.xlabel("Year")
plt.ylabel("Prod")
plt.grid(color="blue", ls ="--", lw=3, alpha= 0.2)
plt.title("YEAR Wise Data")
plt.xticks(x1,['2017','2018','2019','2020','2021'])
plt.show()
```

### 2

```python
data1 = [56,78,334,789,11]
data2 = [67,234,88,23,677]

plt.bar(np.arange(len(data1)), data1, width=0.4)
plt.bar(np.arange(len(data2))+0.4, data2, width=0.4)
plt.show()
```

### 3

```python
data1 = [56,78,334,789,11]
data2 = [67,234,88,23,677]

plt.bar(np.arange(len(data1)), data1)
plt.bar(np.arange(len(data2)), data2, bottom=data1)
```

### 4 histogram

```python

data1= [1,2,3,4,5]
data1 = np.random.normal(170,10,250)
plt.hist(data1)
```

### 5 pie chart

```python

data1= np.array([15,67,23,17,19])
label1= ["prod1","prod2","prod3","prod4","prod1"]
ex1 =[0,0.3,0,0,0]
plt.pie(data1, labels=label1, startangle=180, explode =ex1,
 shadow= True, colors=['r','g','b', 'y', 'gray'])
plt.legend(title=("Year 2021 Production"))
```

### 6 scatter type

```python
data1 = [56,78,334,789,11]
data2 = [67,234,88,23,677]
plt.scatter(data1, data2)
```
